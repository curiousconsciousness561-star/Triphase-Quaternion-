# Triphase Quaternion — Proof Diagrams & Random Distribution Demo

```
import numpy as np, matplotlib.pyplot as plt
from triphase import random_triphase

N = 1000000
q = random_triphase(N)

# Real-part histogram (proof of better uniformity)
plt.figure(figsize=(8,5))
plt.hist(q[:,0], bins=100, density=True, alpha=0.8, color='#ff9900')
plt.title('Triphase Quaternion — Real Part Distribution (n=1e6)')
plt.xlabel('w = real part')
plt.axhline(1, color='red', linestyle='--', label='Uniform (Haar)')
plt.legend()
plt.show()
```
